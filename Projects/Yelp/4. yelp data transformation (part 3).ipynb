#### Correct Datatype & Drop Null Values

The datatype is reviewed and corrected as needed.

```python
# Review datatype
df.info()
```

The binary columns and other integer columns that are of float datatype are likely because some of the rows are NaNs. The proportions of NaNs are reviewed.

```python
df.isna().mean()
```

Some columns have low proportions of NaNs (<0.0001 or <0.01%). These rows are removed from the data set.

```python
# Column names for which NaN are to be dropped.
cols_drop_na = list(df.loc[:3, 'user_review_count':'compliment_photos'].columns)
```

```python
# Column names for which NaN are to be dropped.
cols_drop_na.extend(list(df.loc[:3, ['friend_count']].columns))
cols_drop_na.extend(list(df.loc[:3, ['postal_code']].columns))
cols_drop_na.extend(list(df.loc[:3, 'active life' : 'shopping'].columns))
cols_drop_na.extend(list(df.loc[:3, ['checkin_count']].columns))

# Columns for which rows containing NaNs are to be dropped
cols_drop_na
```

The rows containing NaNs under the above columns are dropped.

```python
# Drop rows containing NaNs under cols_drop_na
df.dropna(axis = 0, how = 'any', subset = cols_drop_na, inplace = True)
```

The *yelping_since* and *postal_code* columns do not need datatype conversions, threfore are removed from the cols_drop_na list.

```python
# Remove column names for which datatype need not be changed.
cols_drop_na.remove('yelping_since')
cols_drop_na.remove('postal_code')
```

```python
# Convert data type to integers
df[cols_drop_na] = df[cols_drop_na].astype('int32')
```

Another column, called 'elite_count' contains upwards of 75% of NaNs. This is deemed significantly high, therefore this column is dropped.

```python
# Drop elite_count
df.drop(columns = 'elite_count', inplace = True)
```

The proportion of the rows where the entire columns associated with the operation hour are null is calculated,

```python
# Proportion of  rows where all columns regarding operation hours are null
df.loc[:,'monday_start_time' : 'saturday_end_time'].isnull().all(axis = 1).sum() / len(df)
```

5% is non-trivial. Since there is no information about the operation hours, the missing values are replaced with the median opening & closing hours.

```python
# Names of columns
business_hour_cols = df.loc[:, 'monday_start_time' : 'saturday_end_time'].columns
business_hour_cols
```

```python
# Create an empty dictionary
median_hour = {}

# For each business hour column, find the median hour
for col in business_hour_cols:
    hours = list(df[col].sort_values(ascending = True).values) # Create a list of hour in ascending order
    median_hour[col] = hours[len(hours) // 2] # Select median from list

median_hour
```

```python
# Fill NaN with median hours where all columns regarding operation hours are NaN
df[df.loc[:,'monday_start_time' : 'saturday_end_time'].isnull().all(axis = 1)] = df.fillna(median_hour)
```

```python
# Show maximum 10 rows
pd.set_option('display.max_rows', 10)
```

Similar to the rows entirely filled with NaNs under business hours, for any row with NaN values for more than one day (i.e., any row more than 2 NaN values under business hours), the NaNs are replaced with median hours.

```python
# Proportion of rows where NaN appears for more than one day of the week
len(df[df[business_hour_cols].isnull().sum(axis = 1) > 2]) / len(df)
```

```python
# Import datetime library 
import datetime

# Replace NaNs with median hours where NaNs appear for more than one day of the week
df[df.loc[:,'monday_start_time' : 'saturday_end_time'].isnull().sum(axis = 1) > 2] = df.fillna(median_hour)
```

Next, it is assumed that, for any row containing NaNs for just one day of the week, that day is closed. For such rows, the NaNs are replaced with 0:00 to indicate no operation.

```python
# Display table
df[business_hour_cols]
```

```python
# Proportion of rows where NaN appears for just one day of the week
len(df[df[business_hour_cols].isnull().sum(axis = 1) == 2]) / len(df)
```

```python
# Replace NaNs with 00:00:00 where NaNs appear for just one day of the week
df.loc[df[business_hour_cols].isnull().sum(axis = 1) == 2, business_hour_cols] = df.fillna(datetime.time(0,0))
```

Now all the columns are reviewed again for any null values. 

```python
# Change settings to show all rows.
pd.set_option('display.max_rows', None)
```

```python
# Review proportions of null values
df.isna().mean()
```

All the null values have been addressed and there is none left. Knowing that there are no null values, the datatypes of some of the columns can now be changed to those more appropriate.

```python
# Changing the datatypes of business rating and restaurant price range to integers.
df[['business_rating', 'restaurantspricerange']] = df[['business_rating', 'restaurantspricerange']].astype('int32')
```

```python
# Changing the datatypes of the 'yelping_since' and 'date' columns to datetimes
df['yelping_since'] = pd.to_datetime(df['yelping_since'])
df['date'] = pd.to_datetime(df['date'])
```
